```python
import pandas as pd
import plotly.express as px
```

# GapMinder Dataset の取得

142各国の人口、寿命、GDP／人を1952年〜2007年にわたって5年ごとに集計したデータセットです。

以下のコードは、このデータセットを取得し、さらに日本のデータを抽出しています。

```python
PLOTLY_DATASETS = 'https://raw.githubusercontent.com/plotly/datasets/master'
GAP_MINDER_URL = f'{PLOTLY_DATASETS}/gapminderDataFiveYear.csv'
GapMinder = pd.read_csv(GAP_MINDER_URL)

Japan = GapMinder.query('country == "Japan"')

GapMinder
```

```python
JKC = GapMinder.query('country in ["Japan", "Korea, Rep.", "China"]')
JKC
```

## 極東３カ国の寿命

```python
px.line(JKC, x='year', y='lifeExp', color='country')
```

## 極東３カ国の人口

## 極東３カ国の GDP/人

```python
px.line(JKC, x='year', y='gdpPercap', color='country')
```

```python
GapMinder2007 = GapMinder.query('year == 2007')

px.scatter (GapMinder2007, x='gdpPercap', y='lifeExp',
            color='continent')
```

データの偏りがあるので、対数グラフで表示してみる。

```python
GapMinder2007 = GapMinder.query('year == 2007')

px.scatter (GapMinder2007, x='gdpPercap', y='lifeExp', log_x=True,
            color='continent')
```

この図では、点がどの国を表しているのか、わからない。ホバー$^\star$で国を確認できるようにしてみよう。

$^\star$ ホバー：マウスポインタをデータに近づけたときの動作のこと

```python
GapMinder2007 = GapMinder.query('year == 2007')

px.scatter (GapMinder2007, x='gdpPercap', y='lifeExp', log_x=True,
            color='continent',
            hover_name='country')
```

人口を点の大きさで表現してみよう（バブルチャート）

```python
px.scatter (GapMinder2007, x='gdpPercap', y='lifeExp', log_x=True,
            size='pop',
            color='continent',
            hover_name='country')
```

## インタラクションを楽しんでみよう

人口が極めて多い国々（中国、インド、米国、インドネシア、ブラジル、日本）しか様子がわからない。人口の表示に上限を設けてみよう。

## 時系列データのアニメーション表示

ここまでは 2007 年のデータだけを用いてきた。

次の例題はデータセットに含まれるすべての年にわたるデータを用いてアニメーション表示する。

※ ここまでの例の表示対象は `GapMinder2007` だったが、以下では `GapMinder` を用いていることに注意。

```python
px.scatter (GapMinder, x='gdpPercap', y='lifeExp', log_x=True,
            size='pop', size_max=40,
            color='continent',
            hover_name='country',
            animation_frame='year', animation_group='country')
```

やってみよう

- スライダーをクリックするとその年のデータを表示することができる。

- スタートボタンをクリックするとアニメーションを表示することができる。

アニメーションの途中から、国々が図の上にはみ出すのが残念 ← 修正しなくちゃ。

```python
px.scatter (GapMinder, x='gdpPercap', y='lifeExp', log_x=True, range_x=[100, 100000], range_y=[25,90],
            size='pop', size_max=40,
            color='continent',
            hover_name='country',
            animation_frame='year', animation_group='country')
```
